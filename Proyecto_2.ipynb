```python
#Grafica de barras para hombres infectados por el ebola en Guinea
Chicos=GN["Boys"][0:-1:20]
Hombres=GN["Adult Male"][0:-1:20]
Ancianos=GN["Old Male"][0:-1:20]
Fechas=GN.index[0:-1:20]
figurita=ltv.figure(figsize=(20,6))
ltv.bar(Fechas,Chicos,color="blue",alpha=0.8,label="Chicos")
ltv.bar(Fechas,Hombres,color="magenta",alpha=0.4,label="Hombres")
ltv.bar(Fechas,Ancianas,color="red",alpha=0.8,label="Ancianos")
ltv.grid()
ltv.title("Distribución del ébola en hombres de Guinea")
ltv.ylabel("Casos Confirmados")
ltv.xticks(rotation=90,fontsize=8)
ltv.legend(loc="upper left",bbox_to_anchor=(1.02,0.5))
ltv.savefig("Hombres_Guinea.png",dpi=300,format="png")
ltv.show()
```

```python
#A continuacion descompondremos las fechas de manera aleatoria, una parte servira para el Machine Learning la otra para la validacion(el machine learning sirve para el analisis predictivo de una base de datos masiva, como en este caso. La validacion sirve para técnica muy utilizada para evaluar los resultados de un análisis estadístico y garantizar que son independientes de la partición entre los datos de entrenamiento y prueba.)
Chicos=GN["Boys"].to_numpy()
GN_Boys=xd.DataFrame({"Boys":Chicos})
GN_Boys.index=xd.to_datetime(GN.index,format="%d/%m/%Y")
steps=50
data_train=GN_Boys[:-steps]
data_test=GN_Boys[-steps:]
print(f"Las fechas de la base de datos es {len(data_train)}")
```

```python
fig,ax=ltv.subplots(figsize=(8,3))
data_train["Boys"].plot(ax=ax,label="train")
data_test["Boys"].plot(ax=ax,label="test")
ax.legend()
ltv.savefig("Hombres_Guinea_cont.png",dpi=300,format="png")
ltv.show()
```

```python
from skforecast.ForecasterAutoreg import ForecasterAutoreg
from sklearn.ensemble import RandomForestRegressor
```

```python
forecaster=ForecasterAutoreg(
  regressor=RandomForestRegressor(random_state=123),
  lags=6
)
```

```python
forecaster.fit(y=data_train["Boys"])
```

```python
#Pronostico de los siguientes 50 datos
steps=50
Pred=forecaster.predict(steps=steps)
```

```python
Pred.index=data_test.index
fig,ax=ltv.subplots(figsize=(8,3))

data_test["Boys"].plot(ax=ax,label="Datos_Reales",lw=6,color="aqua")
Pred.plot(ax=ax,label="Predicciones",lw=2,color="black")
ax.legend()
ltv.grid()
ltv.title("Predicciones_Hombres_Guinea",fontsize=16)
ltv.savefig("Predicciones_Guinea.png",dpi=300,format="png")
ltv.show()
```
