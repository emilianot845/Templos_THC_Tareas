```python
#TAREA 7 TALLER DE HERRAMIENTAS COMPUTACIONALES TEMPLOS VALDEZ JUAN EMILIANO

#EJERCICIO 4 Funcion 3x+7/x**2-4
"""La grafica de la funcion que tomo de base es la siguiente:"""
import numpy as xd
import matplotlib.pyplot as ltv
def f(x):
    return (3*x + 7) / (x**2 - 4)
x_pos = xd.linspace(-10, 10, 2000)
y_pos= f(x_pos)
ltv.plot(x_pos, y_pos, label=r'$\frac{3x + 7}{x^2 - 4}$')
ltv.axvline(x=-2, color='purple', linestyle='--', label='x = -2')
ltv.axvline(x=2, color='black', linestyle='--', label='x = 2')
ltv.xlabel('x')
ltv.ylabel('y')
ltv.title('Funcion de base $\\frac{3x + 7}{x^2 - 4}$')
ltv.legend()
ltv.grid(True)
ltv.show()
```

```python
#EJERCICIO 1 Def de numero primo
def primo(x):
    if x <= 1:
        return "Error"
    elif x == 2:
        return True
    elif x % 2 == 0:
        return False
    else:
        for i in range(3, int(x**0.5) + 1, 2):
            if x % i == 0:
                return False
        return True

def main():
    while True:
        try:
            inc = int(input("Elegir un x>1: "))
            fin = primo(inc)
            print(fin)
            break
        except ValueError:
            print("Error: Entero que satisface la funcion.")

if __name__ == "__main__":
    main()
```

```python
#EJERCICIO 3 FUNCION DIFERENCIAL
import numpy as xd
import matplotlib.pyplot as ltv

def y_dif(x, y):
    return -y + x * y

def res_general(x, C):
    return (x - 1) + C * xd.exp(-x)

val_x = xd.linspace(-2, 5, 100)

ltv.figure(figsize=(8, 6))
ltv.title('Curvas de Funcion dif')
ltv.xlabel('x')
ltv.ylabel('y')

"""C>0"""
for C in [1, 2, 3]:
    val_y = res_general(val_x, C)
    ltv.plot(val_x,val_y, label=f'C > 0, C = {C}', color='cyan')

"""C=0"""
val_y0= res_general(val_x, 0)
ltv.plot(val_x, val_y0, label='C = 0', color='purple')

"""C<0"""
for C_min in [-1, -2, -3]:
    val_y = res_general(val_x, C_min)
    ltv.plot(val_x, val_y, label=f'C < 0, C = {C_min}', color='black')
ltv.legend()
ltv.show()
```

```python
#Suma de pares e impares del 0 al 150
def sum_pares_impares(rango):
    sum_pares = ((rango // 2) * (rango // 2 + 1)) * 2
    sum_impares = ((rango + 1) // 2) ** 2
    return sum_pares, sum_impares
valor_maximo = 150
sum_pares, sum_impares = sum_pares_impares(valor_maximo)
print(f"Suma de nmeros pares del 0 al {valor_maximo}: {sum_pares}")
print(f"Suma de nmeros impares del 0 al {valor_maximo}: {sum_impares}")
```
