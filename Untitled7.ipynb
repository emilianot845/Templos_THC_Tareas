```python
"""A continuacion haremos un histograma"""
import pandas as xd
import matplotlib.pyplot as ltv

link="https://github.com/seismomat/HerramientasComputacionales/blob/main/Aplicaciones/ebola_outbreak_dataset-master/ebola_outbreak_dataset-master/ebola_2014_2016_clean.csv"
code="?raw=true"
url= link+code
df= xd.read_csv(url,sep=",",index_col=1)
df
```

```python
Sen=df[df["Country"]=="Senegal"]
G=Sen['Girls'][0:-1:7]
AF=Sen['Adult Female'][0:-1:7]
OF=Sen['Old Female'][0:-1:7]
Dates=Sen.index[0:-1:7]
barra=ltv.figure(figsize=(12,6))
ltv.bar(Dates,G,color="black",alpha=0.8,label="Niñas")
ltv.bar(Dates,AF,color="green",alpha=0.9,label="Mujeres")
ltv.bar(Dates,OF,color="red",alpha=0.8,label="Ancianas")
ltv.title("Distribucion del ebola en mujeres de Senegal")
ltv.ylabel("Casos confirmados")
ltv.xticks(rotation=90,fontsize=8)
ltv.legend(loc="upper left",bbox_to_anchor=(1.02,0.5))
ltv.savefig("FemaleSen.png",dpi=300,format="png")
ltv.show()
```

```python
B=Sen['Boys'][0:-1:15]
AM=Sen['Adult Male'] [0:-1:15]
OM=Sen['Old Male'] [0:-1:15]
Dates=Sen.index[0:-1:15]
barra=ltv.figure(figsize=(12,6))
ltv.bar(Dates,B,color="black",alpha=0.8,label="Niños")
ltv.bar(Dates,AM,color="green",alpha=0.9,label="Hombres")
ltv.bar(Dates,OM,color="red",alpha=0.8,label="Ancianos")
ltv.title("Distribucion del ebola en hombres de Senegal")
ltv.ylabel("Casos confirmados")
ltv.xticks(rotation=90,fontsize=8)
ltv.legend(loc="upper left",bbox_to_anchor=(1.02,0.5))
ltv.savefig("MaleSen.png",dpi=300,format="png")
ltv.show()
```

```python
Boys=Sen['Boys'].to_numpy()
Sen_Boys=xd.DataFrame({'Boys':Boys})
Sen_Boys.index=xd.to_datetime(Sen.index,format="%d/%m/%Y")
Sen_Boys
```

```python
steps=57
data_train=Sen_Boys[:-steps]
data_test=Sen_Boys[-steps:]
print(f'Las fechas son {len(data_train)}')
```

```python
fig,ax=ltv.subplots(figsize=(8,3))
data_train['Boys'].plot(ax=ax,label='train')
data_test['Boys'].plot(ax=ax,label='test')
ax.legend()
ltv.show()
```
